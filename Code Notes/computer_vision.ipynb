```python
import numpy as np
import cv2

img = cv2.imread("home/gianmaria/01-Data/lena.png")

# Translation moves pixels in an image according to information in the translation matrix.
'''
Generic structure of the translation matrix
1 0 tx
0 1 ty
'''
M = np.float32([1, 0, 200], [0, 1, 50]) # instance of the translation matrix
# N.B.: Translation is done left-to-right and top-to-bottom: to move the other ways, let tx and ty be negative values.

# In a more general way, translation is done through the cv2.warpAffine() function.
new_img = cv2.warpAffine(img, M, (1024, 1024)) # apply the translation define by M on a 1024x1024 image
# N.B.: If the size of the new image is too small, pixels may be "lost" as they would end up out of the picture.
cv2.imshow("Transformed", new_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# Rotation allows to rotate an image according to information given by the rotation matrix.
'''
General structure of the rotation matrix
a b (1 - a)*center.x - b*center.y
-b a b*center.x + (1 - a)*center.y

Normally, the values are taken in the following way:
a = scale * cosAngle
b = scale * sinAngle

The center of rotation acts as the pivot for the transformation.
'''

h,w = img.shape[:2]
M = cv2.getRotationMatrix2D((h // 2, w // 2), 180, 1) # instance of the rotation matrix
# In a more general way, rotation is done through the cv2.warpAffine() function.
new_img = cv2.warpAffine(img, M, (w, h)) # perform the transformation given by M on a picture with the same size as img
# N.B.: Certain scalings/angles result in the loss of data, so be careful when constructing the rotation matrix.
cv2.imshow("Rotated Image", new_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

The *OpenCV* library also supports filtering, which consists in applying some features (such as blurring or sharpening) to an input image by convolution.
In a more general way, convolution is a mathematical operation that uses an input image and a kernel, which is a small (typically odd-dimensioned) matrix containing the weights to apply to the image, in order to produce a new output image.
Keep in mind, however, that convolution is not an invertible operation, meaning that, after applying the operation to an input image, it is not possible to use the output image to go back to the input image.
Most particularly, the filter kernel can be either user-defined or a well known kernel from the *OpenCV* library.

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# Example of a user-defined kernel.
my_kernel = np.array([[1, 0, 1], [1, 0, 1], [1, 0, 1]]) # 3x3 user-defined kernel

# Convolution is now done through the cv2.filter2D() function.
filtered_img = cv2.filter2D(img, -1, my_kernel) # apply the kernel on ALL channels of the image
cv2.imshow("Result", filtered_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

A blurring kernel computes the average value of the image in order to lower its resolution, resulting in the loss of information but, at the same time, in a noise reduction as well (the impact of the filter also depends on the kernel size).
Most blurring kernels (such as blur, box filter or Gaussian blur) use the arithmetic mean, whereas the median kernel employs the median, which comes in handy for denoising.

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/salt_pepper.png")

# The simplest blurring technique is given by the cv2.blur() function.
filtered_img = cv2.blur(img, (3, 3)) # use a 3x3 kernel, resulting in a smaller image with less noise

# A similar result can be obtained using the boxFilter() function.
filtered_img = cv2.boxFilter(img, -1, (3, 3))

# Lastly, the cv2.GaussianBlur() function performs a blurring using a kernel that follows a Gaussian distribution.
filtered_img = cv2.GaussianBlur(img, (3, 3), 1, 1) # use a Gaussian kernel with deviation=1 on both axes

# Alternatively, the cv2.medianBlur() function uses a median filter to remove the "salt and pepper" blur.
filtered_img = cv2.medianBlur(img, 3) # when using the median blur, the kernel MUST be a square matrix

# Showcase the blurred image.
cv2.imshow("Result", filtered_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

A sharpening filter follows the opposite logic of a blurring filter as it tries to increase the image resolution.
While it is possible to perform sharpening by manually computing the corresponding well-known kernel, it is also possible to perform sharpening by the "unsharpening mask" technique, which consists in summing (overlapping) the input image with a blurred copy that only contains information about the colour.

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# The unsharpening mask does a weighted sum between the original image and a blurred copy using the cv2.addWeighted() function.
smoothed_img = cv2.GaussianBlur(img, (9, 9), 10) # using a bigger kernel gets rid of the details and only saves the colours
final_img = cv2.addWeighted(img, 1.5, smoothed_img, -0.5, 0) # add images by considering their importance
# N.B.: The last parameter, known as the gamma parameter, allows to consider colour correction as well.

# Alternatively, just manually input the corresponding well-known kernel and use the cv2.filter2D() function.
sharpen_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]) # well known 3x3 sharpening kernel
final_img = cv2.filter2D(img, -1, sharpen_kernel)

# Showcase the sharpened image.
cv2.imshow("Result", final_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

From a mathematical point of view, it is possible to imagine a contour in an image as a jump in the image's intensity function.
Since these jumps can be quickly spotted through derivatives, it is possible to recover the contour of an image through the Sobel operator, which quantifies changes by taking the derivatives along the x-axis and y-axis of the grayscale image, or through the Laplacian operator, which internally refers to the Sobel operator in order to produce a less noisy image.

```python
import numpy as np
import cv2

# Apply the cv2.cvtColor() function to convert a RGB image into a grayscale image..
img = cv2.imread("/home/gianmaria/01-Data/lena.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# At this point, compute the first derivatives of the image through the cv2.Sobel() function.
grad_x = cv2.Sobel(gray, -1, 1, 0) # derivative for the x-axis
grad_y = cv2.Sobel(gray, -1, 0, 1) # derivative for the y-axis

# Lastly, convert the values of the derivative to be in the range [0, 255] in order to avoid problems with the pixels' values.
# This is done through the cv2.convertScaleAbs() function, which performs both operations in one shot.
abs_x = cv2.convertScaleAbs(grad_x) # convert and scale the x-derivative
abs_y = cv2.convertScaleAbs(grad_y) # convert and scale the y-derivative
grad = cv2.addWeighted(abs_x, 0.5, abs_y, 0.5, 0) # merge the derivatives into a single image

# Showcase the image.
cv2.imshow("Result", grad)
cv2.waitKey(0)
cv2.closeAllWindows()
```

A logical operator for images works in the same way as bit-wise logical operators, starting from a binary mask containing black (unneeded) or white (needed) pixels, which will be overlapped on the input image to extract the information through bit-wise operators (AND, OR, XOR, NOT).

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# Start by creating the binary mask.
mask = np.zeros(img.shape, dtype=np.uint8)
mask = cv2.rectangle(mask, (100, 100), (250, 250), (255, 255, 255), -1)

# AND: A pixel is non-black if and only if both the input and the kernel encode a non-black pixel in that position.
and_img = cv2.bitwise_and(img, mask)
cv2.imshow("Result", and_img)
cv2.waitKey(0)
cv2.closeAllWindows()

# OR: A pixel is black if and only if both the input and the kernel encode a black pixel in that position.
or_img = cv2.bitwise_or(img, mask)
cv2.imshow("Result", or_img)
cv2.waitKey(0)
cv2.closeAllWindows()

# XOR: The input image is trnasformed to a negative image in the "needed" part of the binary mask.
xor_img = cv2.bitwise_xor(img, mask)
cv2.imshow("Result", xor_img)
cv2.waitKey(0)
cv2.closeAllWindows()

# NOT: Flip the bits of the image (mostly used on the mask).
not_img = cv2.bitwise_and(mask)
cv2.imshow("Result", not_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

An image can be easily summarised using a histogram that plots frequencies for each possible pixel value.
This typically comes in handy for adjusting colours.

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

color = ("b","g","r")
for i,col in enumerate(color):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])

# Plot the histogram.
plt.plot(hist,color=col)

# Show the histogram.
plt.xlim([0,256])
plt.show()
```

Histogram equalization consists in normalizing colour brightness in order to increase contrast.

```python
import cv2
import numpy as np 

# Equalize a grayscale image.
img = cv2.imread("/home/gianmaria/01-Data/lena.png", 0) # specifying 0 allows to load the image in grayscale

img_eq = cv2.equalizeHist(img)

show_img = np.hstack([img,img_eq]) # place the grayscale image and the equalized version sideways for comparison

cv2.imshow("Equalization",show_img) #it works well with grey images
cv2.waitKey(0)
cv2.closeAllWindows()
```

```python
import cv2
import numpy as np 

# Equalize a coloured image.
img = cv2.imread("/home/gianmaria/01-Data/lena.png")

eq_channels = []
channels = cv2.split(img)

for ch in channels:
    eq_channels.append(cv2.equalizeHist(ch))

img_eq = cv2.merge(eq_channels)

show_img = np.hstack([img,img_eq]) # place the image and the equalized version sideways for comparison
cv2.imshow("Equalization",show_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

Generally speaking, histogram equalization on the RGB colour space leads to drastic colour changes.
For this reason, it is more convenient to perform equalization on the HSV colour space, which represents pixels by encoding hue, saturation and value.

```python
import cv2
import numpy as np

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# Change the colour space from BGR to HSV (hue, saturation, value)
hsv_img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(hsv_img) # h is the colour, s is the intensity, v is the brightness
eq_v = cv2.equalizeHist(v)

img_eq = cv2.merge([h,s,eq_v])
img_eq = cv2.cvtColor(img_eq, cv2.COLOR_HSV2BGR)

show_img = np.hstack([img,img_eq])
cv2.imshow("equalization",show_img)
cv2.waitKey(0)
```

The main issue of histogram equalization lies in the fact that ot typically results in the loss of details.
Therefore, it is possible to apply adaptive histogram equalization, which repeatedly applies histogram equalization on smaller tiles in the image.
In addition, in order to avoid amplifying noise, it can be convenient to opt for contrast limited adaptive histogram equalization, which also sets a clip limit for contrast.

```python
import cv2
import numpy as np

img = cv2.imread("/home/gianmaria/01-Data/tsukuba.png", 0) # load the grayscale version of the image

# Start by instantiating the CLAHE algorithm.
clahe = cv2.createCLAHE(clipLimit = 2.0, tileGridSize = (8, 8)) # create a CLAHE with clip limit 2 that uses 8x8 tiles
clahe_img = clahe.apply(img) # apply the CLAHE instance to the input image
eq_img = cv2.equalizeHist(img)

show_img = np.hstack([img, eq_img, clahe_img])
cv2.imshow("equalization",show_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

Automatic feature extraction consists in deriving the features (descriptive points and details) of an image.

In a non-binary feature extractor, the descriptor vector is a stream of floats, meaning that points can be compared by Euclidean distance.
Overall, non-binary extractors are slower to use but they are also more robust because they tend to derive the details of the image.

The SIFT feature extractor is able to extract stronger features corresponding to details.
In particular, when using this algorithm, it is possible to observe that, the stronger the extracted feature, the larger the circle representing it.

**N.B.:** There exists a faster, but patented, alternative to the SIFT extractor that is known as SURF.

```python
import cv2
import numpy as np

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# Create the SIFT feature extractor.
f_ex = cv2.SIFT_create()

# Now, compute the descriptor and keypoint arrays.
keypoints, descriptor = f_ex.detectAndCompute(img, None) # it is possible to specify a mask to extract only some keypoints

# Plot the keypoints on the image.
cv2.drawKeypoints(img, keypoints, img, (255, 0, 0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Show the image and keypoints.
cv2.imshow("Keypoints", img)
cv2.waitKey(0)
cv2.closeAllWindows()

# N.B.: Trying to use SURF results in an error because it is patented.
#       However, it is possible to access it using the extra package and setting OPENCV_ENABLE_NONFREE for manual compiling.
```

In a binary feature extractor, the descriptor vector is a binary stream, meaning that points can be compared by Hamming distance.
Overall, binary extractors are faster to use but they tend to be less robust compared to non-binary extractors.

The ORB feature extractor tends to focus on a given number of points to extract and then loops over them, making the extraction algorithm redundant as the same points will be extracted more times.

```python
import cv2
import numpy as np

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

f_ex = cv2.ORB_create(200) # just take 200 points (this helps to reduce redunancy)

# Now, compute the descriptor and keypoint arrays.
keypoints, descriptor = f_ex.detectAndCompute(img, None)

# Plot the keypoints on the image.
cv2.drawKeypoints(img, keypoints, img, (255, 0, 0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Show the image and keypoints.
cv2.imshow("Keypoints", img)
cv2.waitKey(0)
cv2.closeAllWindows()
```
