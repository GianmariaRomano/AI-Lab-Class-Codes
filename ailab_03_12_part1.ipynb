```python
import numpy as np
import cv2

# Recap on translation:
'''
General structure of the translation matrix
1 0 tx
0 1 ty
'''
img = cv2.imread("/home/gianmaria/01-Data/lena.png")

M = np.float32([[1, 0, 200], [0, 1, 50]]) # move by 200 to the right and by 50 downwards
h,w = img.shape[:2]
new_img = cv2.warpAffine(img, M, (w,h))
cv2.imshow("Transformed", new_img)
cv2.waitKey(0)
cv2.closeAllWindows()

# Note: generally speaking, translation is done left-to-right and top-to-bottom through positive inputs, but it is possible to do reverse transformations by setting negative inputs.
A = np.float32([[1, 0, -200], [0, 1, -50]]) # move by 200 to the left and by 50 upwards
h,w = img.shape[:2]
new_img = cv2.warpAffine(img, A, (w,h)) # apply the translation and give an image of the same size as the original one
# cv2.imshow("Transformed", new_img)
# cv2.waitKey(0)
# cv2.closeAllWindows()

# Notice that maintaining the same size during a translation can be problematic, so consider a larger image size to keep all information.
M = np.float32([[1, 0, 200], [0, 1, 50]]) # a specific instance of the translation, which moves by 200 pixels on the x-axis and by 50 pixels on the y-axis
h,w = img.shape[:2]
new_img = cv2.warpAffine(img, M, (1024,1024)) # apply the translation and give a 1024x1024 image
# cv2.imshow("Transformed", new_img)
# cv2.waitKey(0)
# cv2.closeAllWindows()
```

```python
import numpy as np
import cv2

# Rotation: this affine transformation rotates an image with respect to a given pivot through the warpAffine() function.
'''
General structure of the rotation matrix
a b (1 - a)*center.x - b*center.y
-b a b*center.x + (1 - a)*center.y

Normally, the values are taken in the following way:
a = scale * cosAngle
b = scale * sinAngle

The center of rotation acts as the pivot for the transformation.
'''

# Generally speaking, the rotation matrix is created through the getRotationMatrix2D() function, which builds the matrix from the parameters.
img = cv2.imread("/home/gianmaria/01-Data/lena.png")
h,w = img.shape[:2]
M = cv2.getRotationMatrix2D((h // 2, w // 2), 180, 1) # rotation matrix with center (h // 2, w // 2), with angle 180 and scaling factor 1 (no scaling)
new_img = cv2.warpAffine(img, M, (w, h)) # perform the transformation given by M on a picture with the same size as img
cv2.imshow("Rotated Image", new_img)
cv2.waitKey(0)
cv2.closeAllWindows()

# M = cv2.getRotationMatrix2D((h // 2, w // 2), 180, 1) # same as before but with a scaling factor of 2
# new_img = cv2.warpAffine(img, M, (w, h))
# cv2.imshow("Rotated Image", new_img)
# cv2.waitKey(0)
# cv2.closeAllWindows()

# N.B.: Certain scalings/angles result in the loss of data, so be careful when constructing the rotation matrix.
```

All the information about transformations also allows to find the translation matrix given the starting image and the output image, which can be done either using some reference pixels or by (low-level) feature extraction.

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# Create two sets of points to find the mapping.
pts_1 = np.float32([135, 45], [385, 45], [135, 230]) # starting points
pts_2 = np.float32([135, 45], [385, 45], [150, 230]) # ending points

# Now, it is possible to get the transformation matrix M that transforms pts_1 to pts_2 through the dedicated getAffineTransform() matrix.
M = cv2.getAffineTransform(pts_1, pts_2)
new_img = cv2.warpAffine(pts_1, M, (1024, 1024)) # apply the transformation on the set pts_1 described by M
# N.B.: Sometimes, OpenCV causes an error because it is not able to find the affine matrix (typically because the mapping is a persepctive transformation).
```
