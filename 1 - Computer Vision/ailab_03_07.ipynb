The OpenCV library implements standard functions to draw shapes on images.

```python
import numpy as np
import cv2

# Consider an empty image like a canvas and draw some figures in it.
canvas = np.zeros((300, 300, 3), dtype=np.uint8) # create an empty 300x300 image, specifying that it uses 3 channels

# Start creating the colours using the RGB standard, but remembering that OpenCV follows BGR instead.
green = (0, 255, 0)

# The line() function draws a line on the input image going from a starting point to an ending point.
# Recall to match channels as the canvas was defined to be a one-channel image, meaning that only the first element of the colour would be taken.
cv2.line(canvas, (50, 50), (100, 100), green) # draw a green line going from point (50, 50) until point (100, 100)
# cv2.line(canvas, (50, 50), (100, 100), green) # same as before but the line is thicker
cv2.imshow("Draw", canvas)
cv2.waitKey(0)
cv2.closeAllWindows() # only for JupyterLab or Jupyter Notebook
```

```python
import numpy as np
import cv2

canvas = np.zeros((300, 300, 3), dtype=np.uint8)
red = (0, 0, 255)

# It is possible to draw rectangles using the rectangle() function.
cv2.rectangle(canvas, (150, 150), (200, 200), red)
# cv2.rectangle(canvas, (150, 150), (200, 200), red, 10) # same as before but the rectangle is thicker
# cv2.rectangle(canvas, (150, 150), (200, 200), red, -1) # setting thickness = -1 allows to fill the rectangle
cv2.imshow("Draw", canvas)
cv2.waitKey(0)
cv2.closeAllWindows() # only for JupyterLab or Jupyter Notebook
```

```python
import numpy as np
import cv2

canvas = np.zeros((300, 300, 3), dtype=np.uint8)

blue = (255, 0, 0)
x = canvas.shape[1] // 2 # integer division to avoid errors when drawing the circle
y = canvas.shape[0] // 2 # this happens because standard division would return a float

# The circle() function allows to draw a circle by specifying center and radius.
cv2.circle(canvas, (x, y), 10, blue) # draw a blue circle with radius 10 centered at (x, y)
# cv2.circle(canvas, (x, y), 10, blue, 3) # same as before but the circle is thicker
# cv2.circle(canvas, (x, y), 10, blue, -1) # setting thickness = -1 allows to fill teh rectangle
cv2.imshow("Draw", canvas)
cv2.waitKey(0)
cv2.closeAllWindows() # only for JupyterLab or Jupyter Notebook
```

A cool feature of OpenCV is that it allows to split and extract RGB channels from an image.

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png") # load an image

# It is possible to extract the components of an image through the split() function.
b,g,r = cv2.split(img)
channels = np.hstack([b,g,r])

# At this point, it is possible to compare channels using, which allows to show grayscale channels from each component.
cv2.imshow("Channel Comparison", channels) # this allows to compare the corresponding grayscale images
cv2.waitKey(0)
cv2.closeAllWindows()
```

Alternatively, it is possible to perform the same task by remembering that images are basically numpy arrays.
This strategy is better whenever the channels must be extracted iteratively.

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# Extract the components similarly to the previous point.
b,g,r = cv2.split(img)
b = img[:,:,0] # only extract the blue component
channels = np.hstack([b,g,r])
cv2.imshow("Channel Comparison", channels) # this allows to compare the corresponding grayscale images
cv2.waitKey(0)
cv2.closeAllWindows()
```

In addition, it is possible to merge channels as well by means of a dedicated OpenCV function.
Alternatively, it is possible to exploit numpy stacks as well.

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")
b,g,r = cv2.split(img)

# It is possible to merge channel through the merge() function, which will treat the input as a BGR tuple.
new_img = cv2.merge((r, g, b))
cv2.imshow("Merged Image", new_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

Furthermore, it is possible to use image processing images in order to perform geometric transormations as well.
Generally speaking, geometric transformations are represented by matrices, whose dimension allows to define two main types of transformations:
1) Affine transformation (3x2 matrix).
2) Perspective transformation (3x3 matrix): parallel lines are not preserved, so parallel lines may not be still parallel after the transformation.

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# Resizing: this transformation changes the scale of an image either upwards (the image becomes bigger) or downwards (the image becomes smaller).

# Resizing is done by means of the resize() function.
resized_img = cv2.resize(img, (200, 200), interpolation=cv2.INTER_LINEAR) # resize the image to size 200x200
# resized_img = cv2.resize(img, (700, 200), interpolation=cv2.INTER_LINEAR) # resize the image to size 700x200
# resized_img = cv2.resize(img, (200, 700), interpolation=cv2.INTER_LINEAR) # resize the image to size 200x700
# In a more general way, the interpolation algorithm used for the scaling deals with filling missing data during the transformation (not needed for downwards transformations).
cv2.imshow("Transformed", resized_img)
cv2.waitKey(0)
cv2.closeAllWindows()

# Alternatively, it is possible to use the proportion factor instead of the new size to determine how much the image has to be scaled.
resized_img = cv2.resize(img, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR) # resize the image so that both dimensions are halved
cv2.imshow("Transformed", resized_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# Translation: this affine transformation performs a translation through the parameters defined in the dedicated translation matrix.
'''
General structure of the translation matrix
1 0 tx
0 1 ty
'''
M = np.float32([[1, 0, 200], [0, 1, 50]]) # a specific instance of the translation, which moves by 200 pixels on the x-axis and by 50 pixels on the y-axis

# At this point, translation is done by means of the warpAffine() function, which performs the tasks through the information provided by the translation matrix.
h,w = img.shape[:2]
new_img = cv2.warpAffine(img, M, (w,h)) # apply the translation and give an image of the same size as the original one (normally problematic as it can lose information, so it is a better practice to choose a new dimension)

cv2.imshow("Transformed", new_img)
cv2.waitKey(0)
cv2.closeAllWindows()
# Note: generally speaking, translation is done left-to-right and top-to-bottom through positive inputs, but it is possible to do reverse transformations by setting negative inputs.
A = np.float32([[1, 0, -200], [0, 1, -50]]) # same as before but with a new matrix
h,w = img.shape[:2]
new_img = cv2.warpAffine(img, A, (w,h)) # apply the translation and give an image of the same size as the original one

# cv2.imshow("Transformed", new_img)
# cv2.waitKey(0)
# cv2.closeAllWindows()

# Notice that maintaining the same size during a translation can be problematic, so consider a larger image size to keep all information.
M = np.float32([[1, 0, 200], [0, 1, 50]]) # a specific instance of the translation, which moves by 200 pixels on the x-axis and by 50 pixels on the y-axis
h,w = img.shape[:2]
new_img = cv2.warpAffine(img, M, (1024,1024)) # apply the translation and give a 1024x1024 image

# cv2.imshow("Transformed", new_img)
# cv2.waitKey(0)
# cv2.closeAllWindows()
```
