The OpenCV library also supports filtering, which consists in applying some features (such as blurring or sharpening) to an input image by convolution.
In a more general way, convolution is a mathematical operation that uses an input image and a kernel, which is a small (typically odd-dimensioned) matrix containing the weights to apply to the image, in order to produce a new output image via dot product.
Keep in mind, however, that convolution is not an invertible operation, meaning that, after applying the operation to an input image, it is not possible to use the output image to go back to the input image.
Most particularly, it is possible to define several types of convolution:
1) One-dimensional convolution, which is mainly used for signals.
2) Two-dimensional convolution, which is mainly used fo images.
3) Three-dimensional convolution, which is mainly used for medical images as it considers layers and volumes.

Generally speaking, the OpenCV library allows users to create their own kernel through NumPy arrays or through well-known kernels that are designed for dedicated operations, such as (Gaussian) blur or sharpening.

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")


# Start by creating a user-defined kernel.
my_kernel = np.array([[1, 0, 1], [1, 0, 1], [1, 0, 1]]) # 3x3 user-defined kernel

# At this point, it is possible to apply convolution through the filter2D() function.
filtered_img = cv2.filter2D(img, -1, my_kernel) # apply the kernel on ALL channels of the image
cv2.imshow("Result", filtered_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

A blurring kernel acts as an averaging filter because it computes the average of the image, resulting in the loss of information of the image, although this filter can be useful to reduce noise.
The main idea is that, as the kernel size increases, so will the blurring, resulting in a loss of resolution but also in a better denoising.

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/salt_pepper.png")

# The simplest blurring technique is given by the blur() function.
filtered_img = cv2.blur(img, (3, 3)) # apply convolution using a 3x3 kernel, resulting in the creation of a smaller image with less noise
# filtered_img = cv2.blur(img, (7, 7)) # same as before but with a 7x7 kernel, resulting in an image with less noise but less details
cv2.imshow("Result", filtered_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

A sharpening filter follows the opposite logic of a blurring filter as it tries to increase the input image's resolution.
This can be achieved in two main ways
1) Unsharpen mask, which consists in summing the image with a blurred copy.
2) Well-known kernel.

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# A first approach for sharpening consists in performing a weighted sum between the original image and a blurred version through the addWeighted() function.
smoothed_img = cv2.GaussianBlur(img, (9, 9), 10) # using a bigger kernel gets rid of the details and only saves the colours
final_img = cv2.addWeighted(img, 1.5, smoothed_img, -0.5, 0) # add images by considering their importance
# N.B.: The last parameter, known as the gamma parameter, allows to consider colour correction as well.
cv2.imshow("Result", final_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# An alternative approach consists in using the well-known kernel for sharpening.
sharpen_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]) # well known 3x3 sharpening kernel
final_img = cv2.filter2D(img, -1, sharpen_kernel)
cv2.imshow("Result", final_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```

By knowing that the edge of an image can be considered as a jump in the image's intensity function, the following well-known kernels allow to extract the contours of an image:
1) Sobel operator, which considers horizontal and vertical changes through the derivatives for the x-axis and the y-axis on the grayscale image.
2) Laplace operator, which internally calls the Sobel operator, allowing for a faster and less noisy result.

```python
import numpy as np
import cv2

# A first way to turn images into grayscale images is by specifying a flag in the imread() function.
# img = cv2.imread("/home/gianmaria/01-Data/lena.png", cv2.IMREAD_GRAYSCALE) # load a grayscale version of the image (flag = 0)

# Alternatively, apply the cvtColor() function for conversions.
img = cv2.imread("/home/gianmaria/01-Data/lena.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# At this point, compute the first derivatives of the image through the Sobel() function.
grad_x = cv2.Sobel(gray, -1, 1, 0) # derivative for the x-axis
grad_y = cv2.Sobel(gray, -1, 0, 1) # derivative for the y-axis

# Lastly, convert the values of the derivative to be in the range [0, 255] in order to avoid problems with the pixels' values.
# This is done through the convertScaleAbs() function, which performs both operations in one shot.
abs_x = cv2.convertScaleAbs(grad_x) # convert and scale the x-derivative
abs_y = cv2.convertScaleAbs(grad_y) # convert and scale the y-derivative
grad = cv2.addWeighted(abs_x, 0.5, abs_y, 0.5, 0) # merge the derivatives into a single image

cv2.imshow("Result", grad)
cv2.waitKey(0)
cv2.closeAllWindows()
```

A logical operator for images works in the same way as bit-wise logical operators.
In particular, the operator makes use of a binary mask containing black (unneeded) or white (needed) pixels, which will be overlapped on the input image to extract the information through bit-wise operators (AND, OR, XOR and so on and so forth).

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/lena.png")

# Create a binary mask by creating a black image with a white rectangle inside it.
mask = np.zeros(img.shape, dtype=np.uint8)
mask = cv2.rectangle(mask, (100, 100), (250, 250), (255, 255, 255), -1)

# Apply the AND operator, which returns a non-back pixel if and only if both images encode a non-black pixel in that position.
and_img = cv2.bitwise_and(img, mask)
cv2.imshow("Result", and_img)
cv2.waitKey(0)
cv2.closeAllWindows()

# Apply the OR operator, which returns a black pixel if and only if both inputs encode a black pixel in that position.
or_img = cv2.bitwise_or(img, mask)
cv2.imshow("Result", or_img)
cv2.waitKey(0)
cv2.closeAllWindows()

# Apply the XOR operator, which returns a negative image of the input image in the mask.
xor_img = cv2.bitwise_xor(img, mask)
cv2.imshow("Result", xor_img)
cv2.waitKey(0)
cv2.closeAllWindows()

# Apply the NOT operator, which flips the bits of an image (typically the mask).
not_img = cv2.bitwise_and(mask)
cv2.imshow("Result", not_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```
