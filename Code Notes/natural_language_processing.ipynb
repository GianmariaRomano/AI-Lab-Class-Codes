Generally speaking, natural language processing deals with interactions between computers and the human language in order to process and analyze large amount of natural language data. <br>
However, natural language processing models tend to be hard to manage due to the following issues:
- **Ambiguity**: The text may be ambiguous at word/speech/syntax/meaning/language level.
- **Scale**: The chosen model may be able to work with more languages.
- **Sparsity**: Depending on their frequency, some words may be more important than other words.
- **Variation**: The meaning of a sentence may also depend on its context.
- **Expressivity**: The same sentence can be worded in different ways, all with the same meaning.

Tokenizing consists in dividing a text into "tokens", which can be simpler sentences or single words.

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

example_string = "All the speed he took, all the turns he'd taken"

# Apply tokenizing by splitting the text into simpler sentences.
s_tokenized = sent_tokenize(example_string)
print(s_tokenized)

# Apply tokenizing by splitting the text into singe words.
w_tokenized = word_tokenize(example_string)
print(w_tokenized)
```

Generally speaking, a "stop word" is a common word, such as "like" or "in", that, due to its high frequency in texts, tends to provide no useful information for processing the text.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Start by getting the words of the input string.
example_string = "It's leviosa, not leviosaaaa"
words = word_tokenize(example_string)

# Since stop words depend on the language, set the language to English in order to avoid issues when filtering the text.
stop_words = set(stopwords.words("english"))

# At this point, remove the stop words by iterating through the tokenized words.
filtered_list = []
for word in words:
    if word.casefold() not in stop_words: # use the casefold() function to ignore case sensitivity
        filtered_list.append(word)

print(filtered_list)
```

Stemming is a procedure that deals with finding the root of each word in a text.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# Start by getting the words of the input string.
example_string = "The crew of the USS Discovery discovered many discoveries. Discovering is what explorers do"
words = word_tokenize(example_string)

# At this point, create an instance of the stemming model in order to derive the roots of each word in the text.
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in words] # list comprehension performing the stemming task
print(stemmed_words)
```

Word stemming, however, is often subjected to understemming, which happens when related words get reduced to different roots, or overstemming, which happens when unrelated words get reduced to the same root. <br>
Generally speaking, it is possible to handle this issue by applying alternative strategies, such as part of speech tagging, which first determines what part of speech a word represents, or lemmatizing, which instead reduces a word to its core meaning rather than focusing on its root.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.tag import pos_tag

# Start by getting the words of the input string.
example_string = "The crew of the USS Discovery discovered many discoveries. Discovering is what explorers do"
words = word_tokenize(example_string)

# At this point, create an instance of the stemming model in order to derive the root of each word in the text.
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in words]

# Part of speech tagging allows to determine what part of speech each word represents.
tags = nltk.pos_tag(stemmed_words)
print(tags)
# print(nltk.help.upenn_tagset()) # use this command to understand the meaning of each tag
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

# Start by creating an instance of the lemmatizing model.
lemmatizer = WordNetLemmatizer()

# At this point, try using the model on some examples.
x = lemmatizer.lemmatize('scarves')
print(x)

y = lemmatizer.lemmatize('worst', pos='a') # to consider adjectives
print(y)
```
