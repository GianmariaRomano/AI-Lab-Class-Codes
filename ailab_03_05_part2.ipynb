Initially designed for C/C++, OpenCV is the standard Python library for image processing.

```python
import cv2
import numpy as np # for simplicity
import matplotlib.pyplot as plt

# Unlike the images library, the images do not have to be in the same folder as the code.

# It is possible to load images through the imread() function, which always requires the path to the image.
img = cv2.imread("/home/gianmaria/01-Data/lena.png") # nothing is returned as you just read an image

# It is possible to access to image properties through shape.
print(f'The width is {img.shape[1]}') # print the width of the image
print(f'The height is {img.shape[0]}') # print the height of the image
print(f'The channels are {img.shape[2]}') # print the depth, which denotes the number of channels of the image
print(img.shape) # a faster way to print the information of the image

# It is possible to show images on the screen through the imshow() function.
cv2.imshow("Loaded Image", img) # show the image on the "Loaded Image" window (either nothing happens or the windown is opened but closes immediately)
cv2.waitKey(0) # this command allows to show the image for 0 milliseconds -> it is possible to close the window either normally or through a random key on the keyboard
cv2.closeAllWindows() # to avoid fucking things up :/

# It is possible to write an image on the disk through the imwrite() function.
cv2.imwrite("lena.jpg", img) # it will save the input image as lena.jpg

# Since images are treated as numpy arrays, it is possible to create empty images through the zeros() function.
empty = np.zeros((200, 200), dtype=np.uint8) # create an empty 200x200 image
cv2.imwrite("empty.jpg", empty)

# It is possible to access to the value of a pixel by just specifying its position.
b,g,r = img[0, 0] # values of the top left pixel, given in BGR format
print(b,g,r)

# Alternative way of showing an image through matplotlib, allowing to understand that OpenCV follows the BGR paradigm.
# plt.imshow(img)
plt.show()
```
