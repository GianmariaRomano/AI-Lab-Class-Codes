```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/gerry.png")

# Start by defining the starting points and the corresponding destination points, along with the corresponding transformation matrix.
# Starting points (corners of the input image).
src_points = np.float32([[28, 227], [131, 987], [730, 860], [572, 149]]) # anticlockwise order
# Destination points (corners of the resolution of the output image).
dst_points = np.float32([[0, 0], [0, 800], [600, 800], [600, 0]]) # destination points for a 800x600 image
# Get the transformation matrix for scanning.
# M = cv2.getAffineTransform(src_points, dst_points) # NO! Using an affine transformation leads to an error
M = cv2.getPerspectiveTransform(src_points, dst_points)

# Now, perform the transformation to obtain the new image.
# new_img = cv2.warpAffine(img, M, (600, 800)) # NO! Using an affine transformation leads to an error
# Notice that applying an affine transformation will cause an error because four points are used instead of three.
# This task would require a perspective transformation to work with all four corners, so the warpPerspective() function is used instead.
new_img = cv2.warpPerspective(img, M, (600, 800)) # correct command to perform the transformation

# The scanner takes as input the image corners that will be mapped onto the corners (0, 0), (0, col), (row, 0), (row, col) obtained by straightening the image.
cv2.namedWindow("Image", cv2.WINDOW_KEEPRATIO) # create the window before using it so that the window keeps all the mentioned features (in this case, the window keeps ratios)
cv2.imshow("Image", new_img)
cv2.waitKey(0)
cv2.closeAllWindows() # to avoid issues on JupyterLab
```

Alternative version that employs clicking.

```python
import numpy as np
import cv2

img = cv2.imread("/home/gianmaria/01-Data/gerry.png")

# Copy the image for visualization purposes: OpenCV will keep track of chosen points with clicks to make you understand the procedure.
img_copy = img.copy()

# Start by defining the starting points and the corresponding destination points.
# Starting points (corners of the input image): in this version, they will be added at runtime by clicking on the image.
src_points = []
# Destination points (corners of the resolution of the output image).
dst_points = np.float32([[0, 0], [0, 800], [600, 800], [600, 0]]) # destination points for a 800x600 image

# Now, define the callback function (useful for the scanner)
def onClick(event, x, y, flags, params):
    # Only consider clicks done with the left mouse button.
    if event == cv2.EVENT_LBUTTONDOWN:
        # Recall that the perspective transformation only needs four points to work.
        if len(src_points) < 4:
            # Append the clicked point and then mark it on the image with a red circle.
            src_points.append([x, y])
            cv2.circle(img_copy, (x, y), 10, (0, 0, 255), -1) # use the copy to avoid modifying the original image
            cv2.imshow("Image", img_copy)

# Create the window that keeps track of the clicks through a callback function.
cv2.namedWindow("Image", cv2.WINDOW_KEEPRATIO)
cv2.setMouseCallback("Image", onClick) # when the mouse is clicked, do something through the onClick function (defined above at the start)

# Show the image for getting the clicks.
cv2.imshow("Image", img_copy)
cv2.waitKey(0)
cv2.closeAllWindows()

# Get the perspective matrix for the scanning transformation.
src_float = np.float32(src_points) # convert src_points from a standard list to a numpy array to avoid errors at the next command
H = cv2.getPerspectiveTransform(src_float, dst_points)

# Apply the transformation to the original image through the warpPerspective() function and show the final image.
output_img = cv2.warpPerspective(img, H, (600, 800))
cv2.namedWindow("Result", cv2.WINDOW_KEEPRATIO)
cv2.imshow("Result", output_img)
cv2.waitKey(0)
cv2.closeAllWindows()
```
